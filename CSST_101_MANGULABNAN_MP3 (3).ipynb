```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binomial(trials, head):
  return np.random.binomial(n = 1, p = head, size = trials)

def normal(trials, mean, stdev):
  return np.random.normal(loc = mean, scale = stdev, size = trials)

def main():
  choice = input("Choose a distribution to simulate: \n 1. Binomial \n 2. Normal \n")

  if choice == '1':
    trials = int(input("Enter the number of trials: "))
    head = float(input("Enter the probability of heads: "))

    distribution = binomial(trials, head)
    plt.hist(distribution, bins=2, edgecolor='black')
    plt.title('Binomial Distribution (Coin Flips)')
    plt.xlabel('Outcome')
    plt.ylabel('Frequency')
    plt.xticks([0, 1], ['Tails', 'Heads'])

  elif choice == '2':
    trials = int(input("Enter the number of trials: "))
    mean = float(input("Enter the mean for the grade: "))
    stdev = float(input("Enter the standard deviation for the grade: "))
    distribution = normal(trials, mean, stdev)
    plt.hist(distribution, bins=30, edgecolor='black')
    plt.title('Normal Distribution (Exam Scores)')
    plt.xlabel('Score')
    plt.ylabel('Frequency')

  else:
    print("Invalid choice")
    return

  plt.show()

if __name__ == "__main__":
  main()
```
